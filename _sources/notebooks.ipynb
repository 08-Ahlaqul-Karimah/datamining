## Dismilarity

Dalam peambangan data yang pengaplikasian yang sering digunakan adalah clustering. pada teknik ini perlu untuk menemukan seberapa mirip atau tidaknya data. Salah satu cara yaitu dengan menggunakan Dismilarity

## Import Modul

```python
import pandas as pd
import numpy as np
import math 
```

## Get Dataset

```python
dataset_url="https://raw.githubusercontent.com/08-Ahlaqul-Karimah/machine-Learning/main/covid_19_india.csv"
p=pd.read_csv(dataset_url, index_col=0)
```

```python
p
```

## Mencari Dismilarity data nominal

```python
def diss(x,y):
  a = 4
  b = 0
  for col in ["Date", "Time", "State/UnionTerritory", "ConfirmedIndianNational", "",	"ConfirmedForeignNational", "Cured",	"Deaths", "Confirmed"	]:
    if p[col][x] == p[col][y]:
      b+=1
    return (a-b)/a
```

```python
print(diss(1,1))
```

```python
p.describe()
```

```python
#normalisasi Cured
std_cured = 6.148909e+05
mean_cured = 2.786375e+05
for col1 in ["Cured"]:
    p1 = p[col1]
    hasil1 = (p1-mean_cured)/std_cured
    print(hasil1)

#normalisasi Deaths
std_deaths = 10919.076411
mean_deaths = 4062.402264
for col2 in ["Deaths"]:
    p2 = p[col2]
    hasil2 = (p2-mean_deaths)/std_deaths
    print(hasil2)
```

## Menghitung jarak data Numeric dengan cara Euclidean

```python

def num_p(a,b,c):
    result = 0
    for col in [hasil1, hasil2]:
        result = math.pow(abs(col[a] - col[b]), c)
    return result**(1/c)
```

## Manhattan distance

```python
print(num_p(1,2,2))
print(num_p(1,3,2))
print(num_p(1,4,2))
```
